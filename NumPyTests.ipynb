```python
import time

import math

import random

import numpy as np

import matplotlib.pyplot as plt
```

```python
list_tests = list()

n = 1000000

for size in range(100000, 1000000, 5000):
    
    test_list = [random.random() for i in range(size)]
    
    start = time.time()
    
    test_list = [i * n for i in test_list]
    
    list_tests.append(time.time() - start)   
```

```python
np_trivial_tests = list()

n = 1000000

for size in range(100000, 1000000, 5000):
    
    test_array = np.random.random_sample(size)
    
    start = time.time()
    
    time.sleep(0.1)
    
    test_array = test_array * n
    
    np_trivial_tests.append(time.time() - start) 
```

```python
np_dot_tests = list()

n = 1000000

for size in range(100000, 1000000, 5000):
    
    test_array = np.random.random_sample(size)
    
    start = time.time()
    
    time.sleep(0.1)
    
    test_array = test_array.dot(n)
    
    np_dot_tests.append(time.time() - start)
```

```python
%matplotlib inline

plt.figure(figsize=(20, 10))

X = list(range(100000, 1000000, 5000))

plt.plot(X, 
         list_tests, 
         color='r', 
         label='List test')

plt.plot(X, 
         np_trivial_tests,
         color='b',
         label='Trivial NumPy test')

plt.plot(X,
         np_dot_tests,
         color='g',
         label='NumPy.array.dot test')

plt.legend()
plt.grid(True)

plt.xlabel('Size of array')

plt.ylabel('Time of multiplay')

plt.title('Test of NumPy and list in Python')
```

```python
def tanh(x):
    return (math.e**(2*x) - 1) / (math.e**(2*x) + 1)
```

```python
python_func_test = list()

for size in range(100000, 1000000, 5000):
    
    test_list = [random.random() for i in range(size)]
    
    start = time.time()
    
    test_list = [tanh(i) for i in test_list]
    
    python_func_test.append(time.time() - start)  
```

```python
numpy_tanh_test = list()

for size in range(100000, 1000000, 5000):
    
    test_array = np.random.random_sample(size)
    
    start = time.time()
    
    test_array = np.tanh(test_array)
    
    numpy_tanh_test.append(time.time() - start) 
```

```python
plt.figure(figsize=(20, 10))

plt.plot(X, 
         python_func_test,
         color='b',
         label='Python')

plt.plot(X, 
         numpy_tanh_test,
         color='r', 
         label='NumPy')

plt.legend(loc='upper left', fontsize=20)

plt.grid(True)

plt.xlabel('Size of array')

plt.ylabel('Time of work')

plt.title('Hand-made tanh-function vs NumPy.tanh')
```

```python
def normalize(arr):
    x_max = max(arr)
    x_min = min(arr)
    
    return [(i - x_min) / (x_max - x_min) for i in arr]

def np_normalize(arr):
    x_max = arr.max()
    x_min = arr.min()
    
    return (arr - x_min) / (x_max - x_min)
```

```python
py_normalize_test = list()
np_normalize_test = list()

for size in range(100000, 1000000, 5000):
    
    py_array = [random.random() for i in range(size)]
    
    np_array = np.random.random_sample(size)
    
    start = time.time()
    py_array = normalize(py_array)
    py_normalize_test.append(time.time() - start) 
    
    start = time.time()
    np_array = np_normalize(np_array)
    np_normalize_test.append(time.time() - start)    
```

```python
plt.figure(figsize=(20, 10))

plt.plot(X, 
         py_normalize_test,
         color='r', 
         label='Normalize in Python')

plt.plot(X, 
         np_normalize_test,
         color='b',
         label='Normalize with NumPy')

plt.legend(fontsize=20, 
           loc='upper left')

plt.grid(True)

plt.xlabel('Size of array')

plt.ylabel('Time of normalization')

plt.title('Normalization with Python/NumPy')
```
