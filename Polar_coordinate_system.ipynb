```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def spiral():
    
    t = np.arange(0, 8*np.pi, 0.01)
    
    return t, 2 * t
```

```python
def polar_rose():
    
    t = np.arange(0, 2 * np.pi, 0.01)
    
    return t, np.sin(6 * t)
```

```python
def cardioid():
    
    t = np.arange(0, 2 * np.pi, 0.01)
    
    return t, 1 - np.sin(t)
```

```python
def plot(function, title=''):
    
    x, y = function()
    
    plt.clf()
    ax = plt.figure(figsize=(16, 8)).add_subplot(111, projection='polar')

    ax.plot(x, y, linewidth=4, color='blue', label=f'{function.__name__}')
    ax.legend(loc='best', fontsize=16)
    ax.set_title(title)
```

```python
plot(spiral, title='Archimedean spiral')
```

```python
plot(polar_rose, title='Polar rose')
```

```python
plot(cardioid, 'Cardioid')
```
