```python
y_data = df['target_class']

X_data = df.loc[:, :' Skewness of the DM-SNR curve']

X_data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data,
                                                    y_data,
                                                    random_state=42)
X_train.head()
```

```python
tree1 = DecisionTreeClassifier().fit(X_train, y_train)
tree1
```

```python
tree1.score(X_test, y_test)
```

```python
tree1.score(X_train, y_train)
```

```python
tree2 = DecisionTreeClassifier(max_depth=8).fit(X_train, y_train)
tree2.score(X_test, y_test), tree2.score(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_1 = RandomForestClassifier().fit(X_train, y_train)

rf_1.score(X_test, y_test)
```

```python
for n_e, max_depth in [(50, 5), (100, 5), (50, 10), (100, 10)]:
    rf = RandomForestClassifier(n_estimators=n_e, max_depth=max_depth).fit(X_train, y_train)
    print('RandomForest(n_estimators={0}, max_depth={1}) : {2}'.format(n_e,
                                                                       max_depth, 
                                                                       rf.score(X_test, y_test)))
```

```python
%matplotlib inline

import matplotlib.pyplot as plt

fig, axes = plt.subplots(3, 3, figsize=(40, 30))
i = 0

axes = axes.ravel()

for max_depth in range(4, 21, 2):
    n_e = range(25, 76)
    y = []
    for n in n_e:
        rf = RandomForestClassifier(n_estimators=n, max_depth=max_depth).fit(X_train, y_train)
        y.append(rf.score(X_test, y_test))
        
    axes[i].plot(n_e, y, label='Accuracy')
    axes[i].set_title('Max depth is {0}'.format(max_depth))
    axes[i].set_ylabel('Accuracy')
    axes[i].legend()
    
    i += 1 
```

```python
fig.savefig(r'C:\C.png')
```
